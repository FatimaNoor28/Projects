```python
import numpy as np
import random
```

```python
c1 = np.zeros((8,8))
c2 = np.zeros((8,8))
c3 = np.zeros((8,8))
c4 = np.zeros((8,8))
```

```python
for i in range(0,8):
    i1 = random.randrange(0,8)
    i2 = random.randrange(0,8)
    i3 = random.randrange(0,8)
    i4 = random.randrange(0,8)
    c1[i1][i] = 1
    c2[i2][i] = 1
    c3[i3][i] = 1
    c4[i4][i] = 1
```

```python
def findleftDiagonal(c,i,j):
    countdiagonal = 0
    row = i
    for col in range(j,8):
        if row == 8:
            break
        if c[row][col] == 1:
            countdiagonal=countdiagonal+1
        row=row+1
    row = i-1
    for col in range(j-1,-1,-1):
        if row == -1:
            break
        if c[row][col] == 1:
            countdiagonal = countdiagonal+1
        row = row - 1
    return countdiagonal
```

```python
def findrightDiagonal(c,i,j,countdiagonal):
    row = i-1
    for col in range(j+1,8):
        if row == -1:
            break
        if c[row][col] ==1:
            countdiagonal=countdiagonal+1
        row=row-1
    row = i+1
    for col in range(j-1,-1,-1):
        if row == 8:
            break
        if c[row][col] == 1:
            countdiagonal = countdiagonal+1
        row = row + 1
    return countdiagonal
```

```python
def goodness(c):
    QueenNotAttack=0
    countdiagonal=0
    for i in range(0,8):
        countrow = 0
        for j in range(0,8):
            if c[i][j] == 1:
                countdiagonal=0
                countrow= countrow+1
                countdiagonal = findrightDiagonal(c,i,j,findleftDiagonal(c,i,j))
                #print("diagonal count of c[",i,"][",j,"] is",countdiagonal )
            if j==7:
                #print(countrow, countdiagonal)
                if countrow==1 and countdiagonal==1:
                    QueenNotAttack = QueenNotAttack+1

        #if countrow!=0:
            #print("row count" ,countrow)
    print("Queens not attacked in c:",QueenNotAttack)
    return QueenNotAttack
```

```python
def findMax(goodnessc1,goodnessc2,goodnessc3,goodnessc4):
    test = max(goodnessc1,goodnessc2,goodnessc3,goodnessc4)
    print(test)
    if test == goodnessc1:
        keep1 = c1
    elif test == goodnessc2:
        keep1 = c1
    elif test == goodnessc3:
        keep1 = c3
    elif test == goodnessc4:
        keep1 = c4
    return keep1
def find2ndMax(goodnessc1,goodnessc2,goodnessc3,goodnessc4):

    test = max(goodnessc1,goodnessc2,goodnessc3,goodnessc4)
    if test == goodnessc1:
        goodnessc1 = -1
    elif test == goodnessc2:
        goodnessc2 = -1
    elif test == goodnessc3:
        goodnessc3 = -1
    elif test == goodnessc4:
        goodnessc4 = -1
    test = max(goodnessc1,goodnessc2,goodnessc3,goodnessc4)
    print(test)
    if test == goodnessc1:
        keep2 = c1
    elif test == goodnessc2:
        keep2 = c1
    elif test == goodnessc3:
        keep2 = c3
    elif test == goodnessc4:
        keep2 = c4
    return keep2
```

```python
def mutation1x2(m1,m2):
    mutation = np.zeros((8,8))
    for i in range(0,8):
        for j in range(0,4):
            mutation[i][j] = m1[i][j]
    for i in range(0,8):
        for j in range(4,8):
            mutation[i][j] = m2[i][j]
    return mutation
def mutation2x1(m1,m2):
    mutation = np.zeros((8,8))
    for i in range(0,8):
        for j in range(0,4):
            mutation[i][j] = m2[i][j]
    for i in range(0,8):
        for j in range(4,8):
            mutation[i][j] = m1[i][j]
    return mutation
```

```python
def mutation(m):
    x=random.randrange(0,8)
    y = 0
    while(True):
        if y == 8:
            break
        if m[x][y] == 1:
            m[x][y] = 0
        else:
            y=y+1
    y=random.randrange(0,8)
    m[x][y] = 1
    return m


def calculate8Queen(m1,m2,m3,m4,count):
    if count == 5000:
        return -1
    goodness1 = goodness(m1)
    if goodness1 == 8:
        return m1
    goodness2 = goodness(m2)
    if goodness2 == 8:
            return m2
    goodness3 = goodness(m3)
    if goodness3 == 8:
        return m3
    goodness4 = goodness(m4)
    if goodness4 == 8:
        return m4
    keep1 = findMax(goodness1,goodness2,goodness3,goodness4)
    keep2 = find2ndMax(goodness1,goodness2,goodness3,goodness4)
    mutation1 = mutation1x2(keep1,keep2)
    mutation2 = mutation2x1(keep1,keep2)
    mutation1 = mutation(mutation1)
    mutation2 = mutation(mutation2)
    calculate8Queen(keep1,keep2,mutation1,mutation2,count+1)
    return -1
```

```python
EightQueen = calculate8Queen(c1,c2,c3,c4,0)
if EightQueen == -1:
    print("Problem is still unsolved after 5000 iterations")
print(EightQueen)
```
